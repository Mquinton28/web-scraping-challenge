```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
print(soup.prettify())
```

# Scraping
## NASA Mars News Site Title Scraping

```python
results = soup.find_all('div', class_="slide")
```

```python
# Loop through to find title and text elements
for result in results:
    news_titles = result.find('div', class_="content_title")
    news_title = news_titles.find('a').text
    news_ps = result.find('div', class_="rollover_description")
    news_p = news_ps.find('div', class_="rollover_description_inner").text
    
    print("_______________________________________________________________")
    print(news_title)
    print(news_p)
```

## Image Scraping

```python
url = ('https://www.jpl.nasa.gov/spaceimages/')

html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# pull images
images = soup.find_all('a', class_="fancybox")
print(images)
```

```python
featured_img_src = []
for image in images:
    img = image['data-fancybox-href']
    featured_img_src.append(img)
    
featured_img_url = 'https://ww.jpl.nasa.gov' + img
featured_img_url
```

## Mars Hemispheres

```python
mars_hem_img_urls = []
```

### Cerberus Scrape

```python
cerb_url = ('https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced')

response = requests.get(cerb_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
cerb_img = soup.find_all('div', class_='wide-image-wrapper')
print(cerb_img)
```

```python
for img in cerb_img:
    cerb_pic = img.find('li')
    cerb_full = cerb_pic.find('a')['href']
    print(cerb_full)
    mars_hem_img_urls.append(cerb_full)
    
cerb_title = soup.find('h2', class_='title').text
print(cerb_title)
cerb_hem = {'Cerberus Title': cerb_title, 'Cerberus URL': cerb_full}
print('---------------------------------------------------------------------------------------------------------')
print(cerb_hem)
```
